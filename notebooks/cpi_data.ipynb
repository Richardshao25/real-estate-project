```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import all datas

```python
unemployment_data = pd.read_csv('..\\data\\landing\\Unemployment_rate.csv')
cpi_data = pd.read_csv('..\\data\\landing\\cpidata.csv')
housing_interest = pd.read_excel('..\\data\\landing\\Housing_interest_rate.xlsx')
```

```python
housing_interest
```

```python
unemployment_data.dropna(inplace=True)
unemployment_data
```

Extract time columns

```python
unemployment_data['year'] = unemployment_data['Time'].str.split('-').str[1]
unemployment_data['month'] = unemployment_data['Time'].str.split('-').str[0]
```

```python
# change the year to 20xx
unemployment_data['year'] = '20' + unemployment_data['year']
```

```python
# change month to number
unemployment_data['month'] = unemployment_data['month'].replace('Jan', '01')
unemployment_data['month'] = unemployment_data['month'].replace('Feb', '02')
unemployment_data['month'] = unemployment_data['month'].replace('Mar', '03')
unemployment_data['month'] = unemployment_data['month'].replace('Apr', '04')
unemployment_data['month'] = unemployment_data['month'].replace('May', '05')
unemployment_data['month'] = unemployment_data['month'].replace('Jun', '06')
unemployment_data['month'] = unemployment_data['month'].replace('Jul', '07')
unemployment_data['month'] = unemployment_data['month'].replace('Aug', '08')
unemployment_data['month'] = unemployment_data['month'].replace('Sep', '09')
unemployment_data['month'] = unemployment_data['month'].replace('Oct', '10')
unemployment_data['month'] = unemployment_data['month'].replace('Nov', '11')
unemployment_data['month'] = unemployment_data['month'].replace('Dec', '12')
```

```python
unemployment_data
```

```python
yearly_avg = unemployment_data.groupby('year')['Seasonally adjusted (%)'].mean().reset_index()

yearly_avg.columns = ['year', 'Yearly Average Seasonally Adjusted (%)']
```

Export the data into csv 

```python
yearly_avg
yearly_avg.to_csv('..\\data\\curated\\yearly_avg_unemployment.csv', index=False)
```

Process cpi data

```python
cpi_data

# only consider the time when it's december

cpi_data = cpi_data[cpi_data['time'].str.contains('Dec')]

# convert the time to year, which get rid of the month and put 20 in the front
cpi_data["year"] = "20"+cpi_data["time"].str.split('-').str[1]

cpi_data.to_csv('..\\data\\curated\\cpi_data.csv', index=False)
```
